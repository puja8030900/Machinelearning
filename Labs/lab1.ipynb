### The Gapminder provides population data from 1952 to 2007at 5 year intervals for several countries around the world .Compare the population of the european countries Frame United Kingdom Italy Germany and spain over this period using a like chart make appropriate modifications to the chart title ,axis,titles,legend,figure size,fontsize,colors etc to make the chart readable and visually appealling #Line Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px



df = pd.read_csv("D:\machinelearning\MachineLearning2026_PrinceParmar\Datasets\gapminderDataFiveYear.csv")

countries = ["United Kingdom", "Italy", "Germany", "Spain"]
eu_df = df[df["country"].isin(countries)]

plt.figure(figsize=(12, 7))

for country in countries:
    country_data = eu_df[eu_df["country"] == country]
    plt.plot(
        country_data["year"],
        country_data["pop"],
        marker='o',
        linewidth=2,
        label=country
    )

plt.title(
    "Population Growth of European Countries (1952â€“2007)",
    fontsize=16,
    fontweight="bold"
)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Population", fontsize=13)
plt.legend(title="Country", fontsize=11)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()

```

# Project 2

### diamond.csv file contains various attributes like carat, cut, color, clarity, price etcfor over 53,000 diamonds .Visualize the relationship between the carat (Size of diamond) and price using a scatter plot.Instead of using the entire dataset for this visualization, just pick the diamonds with a clarify "SI2" and color "E" .use the values of the "cut" column to color the dots in the scatter plot .make appropriate modifications to the chart title axis, legend,figure size,colors etc. to make the chart and visually appealing - #scatterplot

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("D:\pooja\Machinelearning\Datasets\diamonds.csv")


filtered_df = df[(df["clarity"] == "SI2") & (df["color"] == "E")]


cuts = filtered_df["cut"].unique()


plt.figure(figsize=(12, 7))

for cut in cuts:
    cut_data = filtered_df[filtered_df["cut"] == cut]
    plt.scatter(
        cut_data["carat"],
        cut_data["price"],
        alpha=0.7,
        s=50,
        label=cut
    )

plt.title(
    "Carat vs Price of Diamonds (Clarity: SI2, Color: E)",
    fontsize=16,
    fontweight="bold"
)
plt.xlabel("Carat (Size of Diamond)", fontsize=13)
plt.ylabel("Price (USD)", fontsize=13)
plt.legend(title="Cut Type", fontsize=11)
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
